```python
# pandas 라이브러리는 반드시 import해서 사용한다.
import pandas as pd

# 출력 옵션 설정
pd.set_option('display.max_columns',None) # 모든 컬럼 출력
pd.set_option('display.width',1000) # 출력 너비 지정
pd.set_option('display.unicode.east_asian_width',True) # 한글 정렬 맞추기

# CSV 파일은 공공데이터포털에서 다운로드 받은 자료이다. => encoding='cp949'
# bike02.csv pandas를 이용해서 읽기 
df = pd.read_csv('greenData/bike02.csv',encoding='cp949')

# 상위 데이터 10개 출력
print("상위 10개 데이터")
print(df.head(10))
print()

# 대여건수 1000이상
print('대여건수 1000 이상')
print(df[df['대여건수'] >= 1000])
print()

# 대여건수 내림차순 정렬
sort = df.sort_values(by='대여건수',ascending=False) # 내림차순 정랼 됨
print("대여건수 내림차순 정렬")
print(sort)
print()

# 전체 합계
total = df['대여건수'].sum()
avg = df['대여건수'].mean()
print('전체 합계')
print(total)
print(f"평균 : {avg:,.2f}")
print()

# 대여건수가 가장 많은 날짜
topDate = df.sort_values(by='대여건수',ascending=False)
print("가장 많은 날짜", topDate.iloc[0]['대여일자'])

```

```python
import pandas as pd
df = pd.read_csv('greenData/pm10.csv',encoding='cp949')
# 구분 컬럼의 데이터 중 '평균'인 행을 제외하여 실제 자치구 데이터만 남깁니다. (데이터 전처리)
disc = df[df['구분'] != '평균']
# 미세먼지(PM10) 농도가 25 이상이면서 동시에 초미세먼지(PM25) 농도가 15 이상인 '주의 요망' 지역을 찾습니다.
# 다중 조건 줄 때 and : &, or : **, not : ~
condition = (disc['미세먼지(PM10)']>=25) & (disc['초미세먼지(PM25)'] >=15)
cond02 = disc[condition]
# 조건에 맞는 지역의 구분, 미세먼지(PM10), 초미세먼지(PM25) 컬럼만 선택해 출력하세요.
result = cond02[['구분','미세먼지(PM10)','초미세먼지(PM25)']]
print("--- 미세먼지 및 초미세먼지 주의 지역 ---")
print(result)      
```

```python
# pandas에서 날짜를 추출하는 함수 종류
# CSV에 입력된 날짜는 문자열 데이터 타입이다.
# 날짜형식으로 형변환하는 함수 : datetime()

# 사고일자 컬럼을 datetime 형식으로 변환하고, 
# 이를 활용하여 '사고월'이라는 새로운 파생 변수(컬럼)를 생성하세요.

# 월별 사고 건수를 집계하여 어느 월에 사고가 가장 많이 발생했는지 확인하세요.

# 사망자 수와 부상자 수를 더하여 '사상자수'라는 파생 변수를 생성하세요.

# 노선명별로 그룹화하여 '사상자수'의 총합을 구하고, 
# 사상자가 가장 많이 발생한 상위 5개 노선을 내림차순으로 출력하세요.

# (보너스) 사고 원인이 '졸음'이면서 사망자가 1명 이상 발생한 치명적인 사고는 
# 총 몇 건인지 구하세요.


import pandas as pd
# from datetime import datetime 
df = pd.read_csv('greenData/accident.csv',encoding='cp949')

# 사고일자가 날짜형식으로 변경됨
df['사고일자']=pd.to_datetime(df['사고일자']) # import없이 사용하는 방법
# 사고일자의 월을 출력 
# dt.month:월, dt.year:년, dt.day:일, dt.hour:시,dt.second:초, dt.minute:분
# dt.dayofweek : 요일을 숫자로 반환(0:월요일 ~ 6:일요일)
# dt.day_name() : 요일 이름이 문자열 ('Monday')
# dt.quarter : 분기(1분기 ~ 4분기)
df['사고월'] = df['사고일자'].dt.month 

# 월별 사고 건수를 집계하여 어느 월에 사고가 가장 많이 발생했는지 확인하세요.
# groupby().count() => 데이터가 비어있지 않은 값의 개수를 센다.
# groupby().size() => 그룹별 전체 행의 개수를 센다. (비어있는 값도 개수센다)

# 호준씨 답
# monthCount = df.groupby('사고월').size().sort_values(ascending=False)

#강사 답
monthCount = df.groupby('사고월').size()
# idxmax() => 가장 큰 값이 들어있는 위치(인덱스 이름)에 접근하여 가져옴
maxMonth = monthCount.idxmax()
print(f"월별 사고 발생 건수 : {monthCount}")
print(f"가장 사고가 많은 월 : {maxMonth}월")

#사망자 수와 부상자 수를 더하여 '사상자수'라는 파생 변수를 생성하세요.
df['사상자수'] = df['사망'] + df['부상']

# 노선명별로 그룹화하여 '사상자수'의 총합을 구하고, 
# 사상자가 가장 많이 발생한 상위 5개 노선을 내림차순으로 출력하세요.
result = df.groupby('노선명')['사상자수'].sum().sort_values(ascending=False).head(5)
print(" -- 사상자수 기준 상위 5개 위험 노선 -- ")
print(result)
print()


# (보너스) 사고 원인이 '졸음'이면서 사망자가 1명 이상 발생한 치명적인 사고는 
# 총 몇 건인지 구하세요.
result02 = len(df[(df['원인'] == '졸음') & (df['사망'] >= 1)])
print(f"--- 졸음 운전중 사망자가 발생한 치명적 사고 건수 : {result02}건")
```

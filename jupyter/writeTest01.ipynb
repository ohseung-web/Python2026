```python
import pandas as pd

# 데이터 생성
data = {
    '대여소': ['강남', '강남', '홍대', '홍대', '잠실', '잠실'],
    '이용자수': [120, 150, 80, 90, 200, 180],
    '대여시간': [30, 45, 20, 25, 60, 55],
    '이용일자': ['2024-01-01','2024-01-02','2024-01-01','2024-01-02','2024-01-01','2024-01-02']
}

# DataFrame 생성
df = pd.DataFrame(data)

# 총이용량 컬럼 생성
df['총이용량'] = df['이용자수'] * df['대여시간']
# 1. 대여소별 총이용량 합계
# -------------------------------
total_usage = df.groupby('대여소')['총이용량'].sum()
print("대여소별 총이용량 합계")
print(total_usage)
print()

# -------------------------------
# 2. 가장 바쁜 날짜 찾기
# -------------------------------
busy_date = df.groupby('이용일자')['이용자수'].sum()
busy_date = busy_date.sort_values(ascending=False)

print("가장 바쁜 날짜")
print(busy_date)
print("가장 바쁜 날짜:", busy_date.index[0])
print()

# -------------------------------
# 3. 평균 이용자수 가장 높은 대여소
# -------------------------------
avg_users = df.groupby('대여소')['이용자수'].mean()
avg_users = avg_users.sort_values(ascending=False)

print("대여소별 평균 이용자수")
print(avg_users)
print("평균 이용자수 가장 높은 대여소:", avg_users.index[0])
```

```python
import pandas as pd

# 1. CSV 파일 읽기
df = pd.read_csv('greenData/bike.csv', encoding='cp949')

# 2. 상위 10개 출력
print("상위 10개 데이터")
print(df.head(10))
print()

# 3. 날짜 데이터 변환
df['대여일시'] = pd.to_datetime(df['대여일시'])

# 4. 이용시간이 3 이상인 데이터만 출력
print("이용시간 3 이상")
print(df[df['이용시간'] >= 3])
print()

# 5. 이용거리 기준 내림차순 정렬
print("이용거리 내림차순 정렬")
sorted_df = df.sort_values(by='이용거리(M)', ascending=False)
print(sorted_df)
print()

# 6. 날짜 컬럼 추가 (날짜만 추출)
df['대여날짜'] = df['대여일시'].dt.date

# 7. 날짜별 총 이용시간
date_usage = df.groupby('대여날짜')['이용시간'].sum()
print("날짜별 총 이용시간")
print(date_usage)
print()

# 8. 가장 이용시간이 많은 날짜
date_usage_sorted = date_usage.sort_values(ascending=False)
print("가장 이용시간이 많은 날짜")
print(date_usage_sorted.index[0])
print()

# 9. 시간 컬럼 추가 (시 단위)
df['대여시간(시)'] = df['대여일시'].dt.hour

# 10. 시간대별 평균 이용거리
hour_avg = df.groupby('대여시간(시)')['이용거리(M)'].mean()
hour_avg_sorted = hour_avg.sort_values(ascending=False)

print("평균 이용거리 높은 시간대")
print(hour_avg_sorted)
print("가장 높은 시간:", hour_avg_sorted.index[0])
print()

# 11. 이용거리 가장 긴 기록 (TOP 1)
top = df.sort_values(by='이용거리(M)', ascending=False)

print("이용거리 TOP 1")
print(top.iloc[0])
```

```python
import pandas as pd

# 1. 출력 옵션 설정 (데이터프레임이 잘리지 않고 예쁘게 나오도록)
pd.set_option('display.max_columns', None)  # 모든 컬럼 출력
pd.set_option('display.width', 1000)        # 출력 너비 설정
pd.set_option('display.unicode.east_asian_width', True) # 한글 정렬 맞추기

# 1. CSV 파일 읽기
df = pd.read_csv('greenData/bike02.csv', encoding='cp949')

# 2. 상위 5개 출력
print("상위 5개 데이터")
print(df.head())
print()

# 3. 대여건수 1000 이상
print("대여건수 1000 이상")
print(df[df['대여건수'] >= 1000])
print()

# 4. 대여건수 내림차순 정렬
sorted_df = df.sort_values(by='대여건수', ascending=False)
print("대여건수 내림차순")
print(sorted_df)
print()

# 5. 전체 합계
total = df['대여건수'].sum()
print("전체 대여건수 합계:", total)
print()

# 6. 평균
avg = df['대여건수'].mean()
print("평균 대여건수:", avg)
print()

# 7. 가장 많은 날짜
top = df.sort_values(by='대여건수', ascending=False)
print("가장 많은 날짜:", top.iloc[0]['대여일자'])
```

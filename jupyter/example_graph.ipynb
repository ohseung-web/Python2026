#### 과목별 점수를 세로 막대그래프로 표현하시오 
#### 그래프 제목: "과목별 점수" 
#### x축: 과목 / y축: 점수 
#### 막대 위에 점수 값 표시
#### python 항목 부분에만 레이블 값이 출력되도록 수정하시오.

```python
# 반드시 pandas는 import하여 사용한다.
import pandas as pd
# matplotlib 매트플롯립도 반드시 import하여 사용한다.
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "과목": ["Java", "Python", "React", "SQL"],
    "점수": [85, 90, 75, 80]
})

# ---------------------------
# 4. 한글 깨짐 방지 설정
# ---------------------------
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 그래프 크기 지정 : figure(10,6)
plt.figure(figsize=(8,5))

bars = plt.bar(df['과목'],df['점수'],width=0.4)


# for range(len())
for i in range(len(bars)):
    bar = bars[i]
    he = bar.get_height() # 세로막대 높이 값

    # python인 경우만 출력
    if df['과목'][i] == 'Python':
        plt.text(bar.get_x()+bar.get_width()/2,he,str(he),ha='center',va='bottom')

plt.title('과목별 점수')
plt.xlabel('과목')
plt.ylabel('점수')

plt.show()
    
```
